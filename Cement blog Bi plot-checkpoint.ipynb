```python
#rename columns for simplification
names = ['Cement', 'Blast_Furnace','Fly_Ash','Water','Superplasticizer',\
         'Coarse_Aggregate','Fine_Aggregate','Age','compressive_strength']
```

```python
data = pd.read_excel(url,names=names)
data.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
X=data.values
```

```python
X=data[['Cement', 'Blast_Furnace','Fly_Ash','Water','Superplasticizer','Coarse_Aggregate','Fine_Aggregate','Age']].values
y=data[['compressive_strength']].values
```

```python
scaler = StandardScaler()
scaler.fit(X)
X=scaler.transform(X)    

pca = PCA()
x_new = pca.fit_transform(X)
```

```python

def myplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.plot([0.66,0.66],[0.5,-0.5],color='m')
    plt.plot([-0.66,-0.66],[0.5,-0.5],color='m')
    
    plt.plot([0.5,-0.5],[0.66,0.66],color='m')
    plt.plot([0.5,-0.5],[-0.66,-0.66],color='m')
    
    plt.plot([0.4375,0.4375],[0.5,-0.5],color='c')
    plt.plot([-0.4375,-0.4375],[0.5,-0.5],color='c')
    
    plt.plot([0.5,-0.5],[-0.4375,-0.4375],color='c')
    plt.plot([0.5,-0.5],[0.4375,0.4375],color='c')
    
    plt.scatter(xs * scalex,ys * scaley,c=y)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.7)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'k', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'k', ha = 'center', va = 'center')

    
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel("PC{}".format(1))
plt.ylabel("PC{}".format(2))

plt.figure(figsize=(15,8)) 
plt.grid()
```

```python
lable=['Cement', 'Blast_Furnace','Fly_Ash','Water','Superplasticizer','Coarse_Aggregate','Fine_Aggregate','Age']
```

```python
myplot(x_new[:,0:2],np.transpose(pca.components_[0:2, :]),labels=lable)

plt.show()
```
